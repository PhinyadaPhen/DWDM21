## **ข้อสอบที่ต้องทำ**

````python
Merge_flights_airports = flights.merge(airports,how='left',left_on='รหัสสนามบินต้นทาง',right_on='รหัสสนามบิน') # รวมตารางโดยใช้ Attribute คือ รหัสสนามบินต้นทาง กับ รหัสสนามบิน
Merge_flights_airports
````

````python
Merge_flights_airports.dropna() # ดูค่า missing
````

### ข้อ 2 วาด Histogram แสดงจำนวนเครื่องบินขาออกของแต่ละวันในสัปดาห์

````python
import pandas as pd
import matplotlib.pyplot as plt # import matplotlib เพื่อเรียกใช้งานฟังก์ชั่นสำหรับพล็อตฮิสโทแกรม
````

### ข้อ 3 สายการบินไหนมีเที่ยวบินมากที่สุดและน้อยที่สุด

````python
Merge_flights_airports.groupby('สายการบิน').count() # นับจํานวนรายการในคอลัมน์
````

````python
FlightsNew = Merge_flights_airports.groupby('สายการบิน').count().sort_values(by=['วันในเดือน'],ascending=False)
FlightsNew.iloc[:,0:1] # ตัดตารางเพื่อให้ดูข้อมูลง่ายขึ้น
````

````python
MaxFlight = FlightsNew.iloc[:1,:1] # สายการบินที่มีเที่ยวบินมากที่สุด
MaxFlight
````

ดังนั้น สายการบินที่มีเที่ยวบินมากที่สุด คือ สายการบิน WN

````python
MinFlight = FlightsNew.iloc[15:,:1] # สายการบินที่มีเที่ยวบินน้อยที่สุด
MinFlight
````

ดังนั้น สายการบินที่มีเที่ยวบินน้อยที่สุด คือ สาบการบิน HA

### โจทย์คิดเอง เมืองใดมีจำนวนเครื่องบินขาออกมากที่สุด

````python
Merge_flights_airports.groupby('เมือง').count() # นับจำนวนรายการในคอลัมน์
````

````python
City = Merge_flights_airports.groupby('เมือง').count().sort_values(by=['ความล่าช้าขาออก'],ascending=False)
City.iloc[:,5:6] # ตัดตารางที่ไม่ใช้ออกเพื่อให้ดูขอมูลง่ายขึ้น
````

````python
MaxCity = City.iloc[:1,:1] # สายการบินที่มีเที่ยวบินมากที่สุด
MaxCity
````

สรุป เมืองที่มีสายการบินขาออกมากที่สุดคือ เมือง Chicago
